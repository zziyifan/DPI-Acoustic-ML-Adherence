```python
# Adjust the text labels
df = pd.DataFrame(data_set_6)
plt.rcParams.update({'font.size': 20})

fig, axes = plt.subplots(1, 2, figsize=(16, 10))
sns.barplot(x='Algorithms', y='best_cv_accuracy_mean', data=df, ax=axes[0], palette='muted')
axes[0].set_title('Best CV Accuracy Mean')
axes[0].set_ylabel('Values')
axes[0].set_ylim([0,1])
axes[0].tick_params(axis='x', labelrotation=45)
axes[0].set_xticklabels(axes[0].get_xticklabels(), ha='right')

# Add values below the bar top 
for p in axes[0].patches:
    axes[0].annotate(format(p.get_height(), '.4f'), 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha = 'center', va = 'center', 
                     xytext = (0, -12), 
                     color='white',
                     textcoords = 'offset points')

sns.barplot(x='Algorithms', y='test_accuracy', data=df, ax=axes[1], palette='muted')
axes[1].set_title('Test Accuracy')
axes[1].set_ylabel('Values')
axes[1].set_ylim([0,1])
axes[1].tick_params(axis='x', labelrotation=45)
axes[1].set_xticklabels(axes[0].get_xticklabels(), ha='right')

# Add values below the bar top
for p in axes[1].patches:
    axes[1].annotate(format(p.get_height(), '.4f'), 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha = 'center', va = 'center', 
                     xytext = (0, -12), 
                     color='white',
                     textcoords = 'offset points')

plt.tight_layout()
plt.show()
```
